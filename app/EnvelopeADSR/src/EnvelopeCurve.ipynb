```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# attack
_maxLevel = 4096
_attack = 100
ex_ = []
for i in range(_attack):
    _level = _maxLevel - (math.exp(-1 * i * (2.0 / _attack)) * _maxLevel)
    ex_.append(_level)
print(ex_[-1])
plt.plot(ex_)
plt.grid(True)
plt.show()
```

```python
# attack 2
_maxLevel = 4096
_attack = 2345
ex_ = []
for i in range(_attack):
    _level = (math.exp(i * (10.0 / _attack)) / math.exp(10.0))
    ex_.append(_level)
print(ex_[-1])
plt.plot(ex_)
plt.grid(True)
plt.show()
_attack * (10.0 / _attack)
```

```python
# decay
_maxLevel = 4096
_sustain = 2000
_decay = 200
ex_ = []
for i in range(_decay):
    _level = (math.exp(-1 * i * (5.0 / _decay)) * (_maxLevel - _sustain)) + _sustain
    ex_.append(_level)
print(ex_[-1])
plt.plot(ex_)
plt.grid(True)
plt.show()
```

```python
# release
_maxLevel = 4096
_sustain = 2000
_release = 200
ex_ = []
for i in range(_release):
    _level = (math.exp(-1 * i * (4.0 / _release)) * (_sustain))
    ex_.append(_level)
print(ex_[-1])
plt.plot(ex_)
plt.grid(True)
plt.show()
```
